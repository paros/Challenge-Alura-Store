### Importación de datos



```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingresos_Tienda1 = tienda["Precio"].sum()
ingresos_Tienda2 = tienda2["Precio"].sum()
ingresos_Tienda3 = tienda3["Precio"].sum()
ingresos_Tienda4 = tienda4["Precio"].sum()

print(f'El valor total de la tienda 1 es: ${ingresos_Tienda1}')
print(f'El valor total de la tienda 2 es: ${ingresos_Tienda2}')
print(f'El valor total de la tienda 3 es: ${ingresos_Tienda3}')
print(f'El valor total de la tienda 4 es: ${ingresos_Tienda4}')

import matplotlib.pyplot as plt

tiendas =['Tienda1', 'Tienda2','Tienda3','Tienda4']
valores_totales = [ingresos_Tienda1,ingresos_Tienda2,ingresos_Tienda3,ingresos_Tienda4]

plt.figure(figsize=(10, 6))
plt.bar(tiendas, valores_totales, color=['blue','green','yellow','purple'])
plt.title('Ingresos totales por tienda')
plt.xlabel('Tiendas')
plt.ylabel('Ingresos totales($)')
plt.show()
```

# 2. Ventas por categoría

```python
VPC_tienda1 = tienda["Categoría del Producto"].value_counts()
VPC_tienda2 = tienda2["Categoría del Producto"].value_counts()
VPC_tienda3 = tienda3["Categoría del Producto"].value_counts()
VPC_tienda4 = tienda4["Categoría del Producto"].value_counts()

VTotal_categoria = tienda.groupby('Categoría del Producto')[['Precio']].sum().sort_values('Precio', ascending = False)
VTotal_categoria2 = tienda2.groupby('Categoría del Producto')[['Precio']].sum().sort_values('Precio', ascending = False)
VTotal_categoria3 = tienda3.groupby('Categoría del Producto')[['Precio']].sum().sort_values('Precio', ascending = False)
VTotal_categoria4 = tienda4.groupby('Categoría del Producto')[['Precio']].sum().sort_values('Precio', ascending = False)

import pandas as pd
tabla = pd.DataFrame({
    'Tienda 1': VPC_tienda1,
    'Tienda 2': VPC_tienda2,
    'Tienda 3': VPC_tienda3,
    'Tienda 4': VPC_tienda4
})
print(tabla,"\n\n")

tabla2 = pd.DataFrame({
    'Tienda 1': VTotal_categoria['Precio'],
    'Tienda 2': VTotal_categoria2['Precio'],
    'Tienda 3': VTotal_categoria3['Precio'],
    'Tienda 4': VTotal_categoria4['Precio']
})
print(tabla2)

```

# 3. Calificación promedio de la tienda


```python
Calificacion_Tienda1 = tienda["Calificación"].mean()
Calificacion_Tienda2 = tienda2["Calificación"].mean()
Calificacion_Tienda3 = tienda3["Calificación"].mean()
Calificacion_Tienda4 = tienda4["Calificación"].mean()
```

```python
print(f'La calificación promedio de la tienda 1 es de: {Calificacion_Tienda1:.2f}')
print(f'La calificación promedio de la tienda 2 es de: {Calificacion_Tienda2:.2f}')
print(f'La calificación promedio de la tienda 3 es de: {Calificacion_Tienda3:.2f}')
print(f'La calificación promedio de la tienda 4 es de: {Calificacion_Tienda4:.2f}')

tiendas =['Tienda', 'Tienda2','Tienda3','Tienda4']
calificacion_promedio = tienda['Calificación'].mean()
calificacion_promedio2 = tienda2['Calificación'].mean()
calificacion_promedio3 = tienda3['Calificación'].mean()
calificacion_promedio4 = tienda4['Calificación'].mean()

volores_totales_calificacion = [calificacion_promedio,calificacion_promedio2,calificacion_promedio3,calificacion_promedio4]

plt.figure(figsize=(10, 6))
plt.pie(volores_totales_calificacion, labels=tiendas, autopct='%1.1f%%', colors=['BLUE','GREEN','YELLOW','PURPLE'])
plt.title('Distribución de Calificación Promedio por Tienda')
plt.axis('equal')
plt.show()
```

# 4. Productos más y menos vendidos

```python
Productos_tienda1 = tienda["Producto"].value_counts()
Productos_tienda2 = tienda2["Producto"].value_counts()
Productos_tienda3 = tienda3["Producto"].value_counts()
Productos_tienda4 = tienda4["Producto"].value_counts()

productos = pd.DataFrame({
    "Más vendido": [
        tienda['Producto'].value_counts().idxmax(),
        tienda2['Producto'].value_counts().idxmax(),
        tienda3['Producto'].value_counts().idxmax(),
        tienda4['Producto'].value_counts().idxmax()
    ],
    "Menos vendido": [
        tienda['Producto'].value_counts().idxmin(),
        tienda2['Producto'].value_counts().idxmin(),
        tienda3['Producto'].value_counts().idxmin(),
        tienda4['Producto'].value_counts().idxmin()
    ]
}, index=["Tienda 1","Tienda 2","Tienda 3","Tienda 4"])

print(productos)
```

Cálculo de precio promedio por tienda y desviación estándar del precio:

```python
precio_T1 = tienda["Precio"].mean()
precio_T2 = tienda2["Precio"].mean()
precio_T3 = tienda3["Precio"].mean()
precio_T4 = tienda4["Precio"].mean()

std_T1 = tienda["Precio"].std()
std_T2 = tienda2["Precio"].std()
std_T3 = tienda3["Precio"].std()
std_T4 = tienda4["Precio"].std()
```

```python
#Tabla Comparativa
analisis = pd.DataFrame({
    "Precio Promedio": [precio_T1, precio_T2, precio_T3, precio_T4],
    "Desviacion Precio": [std_T1, std_T2, std_T3, std_T4]
},
index=["Tienda 1","Tienda 2","Tienda 3","Tienda 4"])

analisis["Coef_Variacion"] = (
    analisis["Desviacion Precio"] /
    analisis["Precio Promedio"]
)

analisis
```

#Informe Final
El presente análisis tuvo como objetivo evaluar el desempeño comercial de cuatro tiendas con el propósito de identificar posibles diferencias en su rendimiento y determinar si estas permiten sustentar una decisión estratégica informada.

Para ello, se analizaron variables clave tales como el ingreso total por tienda, el ingreso promedio por venta, el liderazgo en categorías de productos, el nivel promedio de satisfacción de los clientes y la estabilidad del modelo de precios. A partir de los resultados obtenidos, se observa que no existe una relación directa entre el nivel de satisfacción promedio de los clientes y el ingreso total generado por cada tienda. En consecuencia, la satisfacción del cliente no explica las diferencias observadas en el desempeño comercial.

Asimismo, considerando que la cantidad total de ventas es equivalente entre las tiendas, las diferencias en los ingresos se explican principalmente por el ingreso promedio obtenido por cada venta.

Por lo tanto, se identifican los siguientes hallazgos:

- La Tienda 1, presenta el mayor ingreso promedio por venta, lo que explica su mayor ingreso total.

- La Tienda 2 muestra un desempeño relativamente bajo en tres de las cuatro categorías con mayor volumen de ventas.

- Las Tiendas 3 y 4 lideran ciertas categorías específicas, aunque sin una dominancia clara o sostenida.

- El costo promedio de envío es mayor en la tienda 1

Adicionalmente, el análisis de la variabilidad mediante el coeficiente de variación permite evaluar la estabilidad del modelo; por lo tanto, una decisión estratégica debería considerar de manera conjunta los siguientes factores: el nivel de ingreso promedio por venta, el liderazgo en categorías de productos y la estabilidad del modelo de precios.

La tienda 4 es la que presenta menor ingreso promedio por venta y mayor variabilidad en su desempeño por lo tanto es la menos eficiente.
